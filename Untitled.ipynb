```python
import torch
from torch import nn,optim
from torch.autograd import Variable
from torch.nn import init
from matplotlib import pyplot as plot
import pandas as pd
import numpy as np
```

```python
csvfile = pd.read_csv("../steering.csv")
```

```python
print(csvfile.head(10))
```

```python
plot.figure(figsize=(30,16))
plot.plot(csvfile.timestamp, csvfile.angle,label='angle')
plot.xlabel("time")
plot.plot(csvfile.timestamp, csvfile.torque,label="torque")
plot.plot(csvfile.timestamp, csvfile.speed, label='speed')
plot.legend()
```

```python
angle = csvfile.angle
torque = csvfile.torque
speed = csvfile.speed
```

```python
print(csvfile.timestamp[200]-csvfile.timestamp[150])
```

```python
timestamp = pd.to_datetime(csvfile.timestamp)
```

```python
print(timestamp[0],timestamp[50])
```

```python
data = np.vstack((angle, torque, speed)).T
```

```python
data.shape
```

```python
train_data = data[0:len(data)-1]
train_label = data[1:len(data)]
```

```python
class rnnModel(nn.Module):
    def __init__(self,in_dim,hidden_dim,out_dim,layer_num):
        super().__init__()
        self.rnnLayer=nn.RNN(in_dim,hidden_dim,layer_num)
        self.fcLayer=nn.Linear(hidden_dim,out_dim)
        optim_range=np.sqrt(1.0/hidden_dim)
        self.weightInit(optim_range)

    def forward(self, x):
        out,_=self.rnnLayer(x)
        out=self.fcLayer(out)
        return  out

    def weightInit(self,gain=1):
        for name ,param in self.named_parameters():
            if 'rnnLayer.weight' in name:
                init.orthogonal(param,gain)
```

```python
len(train_data)
```

## One sample

```python
print("the order of ouput data is angle, torque, speed")
rnn_one=rnnModel(3,10,3,2)
criterion_one=nn.MSELoss()
optimizer_one=optim.Adam(rnn_one.parameters(),lr=1e-3)
sec = 200
initial_data = torch.from_numpy(train_data[0].astype(np.float32)).reshape(1,1,3)
loss_one = []
for e in range(0, len(train_data)) :
    inputs = Variable(initial_data)
    target = Variable(torch.from_numpy(train_label[e].astype(np.float32)).reshape(1,1,3))
    # forward
    output = rnn_one(inputs)
    loss = criterion_one(output, target)
    # update gradients
    optimizer_one.zero_grad()
    loss.backward()
    optimizer_one.step()
    # print training information
    print_loss = loss.item()
    loss_one.append(print_loss)
    if e % sec == 0 :
        print('Epoch[{}/{}], predict data is {:.5f} {:.5f} {:.5f}, groud truth is {:.5f} {:.5f} {:.5f} loss = {:.10f}'.format(e,len(train_data),output.detach().numpy()[0,0,0],output.detach().numpy()[0,0,1],output.detach().numpy()[0,0,2],train_label[e,0],train_label[e,1],train_label[e,2], print_loss))
    initial_data = output
```

```python
plot.figure(figsize=(30,50))
plot.ylim(0,1)
plot.plot(range(0, len(train_data)), loss_one)
plot.xlabel("steps")
plot.ylabel("loss")
```

```python
min(loss_one)
```
